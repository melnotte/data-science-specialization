#BASE DE DATOS LOCAL

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
# Creando el motor SQLite
engine = create_engine('sqlite:///:memory:')
```

```python
file = 'clientes_banco.csv'
```

```python
clientes = pd.read_csv(file)
clientes.head()
```

```python
clientes.shape
```

## CREATE

```python
clientes.to_sql('clientes', con=engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renta = "Empleado"'
```

```python
empleados = pd.read_sql_query(sql=text(query), con=engine.connect())
```

```python
type(empleados)
```

```python
empleados
```

```python
empleados.to_sql('empleados', con=engine.connect(), index=False,)
```

## READ


```python
pd.read_sql_table('empleados', con=engine.connect(), columns=['ID_Cliente','Grado_estudio','Rendimiento_anual'])
```

## DELETE


```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(sql=text(query), con=engine.connect())
```

```python
from sqlalchemy.exc import  SQLAlchemyError
query = 'DELETE FROM clientes WHERE ID_Cliente = 5008804'
try:
  r_set = engine.connect().execute(text(query))

except SQLAlchemyError as e:
    print(e)
else:
  print("#Records deleted: ", r_set.rowcount)
```

```python
pd.read_sql_table('clientes', con=engine.connect())
```

## UPDATE

```python
from sqlalchemy.exc import  SQLAlchemyError
query = 'UPDATE clientes SET Grado_estudio = "Nivel superior" WHERE ID_Cliente = 5008806'
try:
  r_set = engine.connect().execute(text(query))

except SQLAlchemyError as e:
    print(e)
else:
  print("#Records updated: ", r_set.rowcount)
```

1. Actualizar el registro del cliente ID 6840104 cuyo rendimiento anual cambió a 300000.


```python
query = 'SELECT ID_Cliente, Rendimiento_anual FROM clientes_banco WHERE ID_Cliente = 6840104'
```

```python
cliente = pd.read_sql_query(sql=text(query), con=engine2.connect())
cliente
```

```python
query = 'UPDATE clientes_banco SET Rendimiento_anual = 300000 WHERE ID_Cliente = 6840104'
try:
  r_set = engine2.connect().execute(text(query))

except SQLAlchemyError as e:
    print(e)
else:
  print("#Records updated: ", r_set.rowcount)
```
