# Visualización y Slicing

```python
# Número de meses
dates = np.arange(1,88,1) # inicio, fin(¡no inclusivo!), paso
dates
```

```python
# Precios de cada mes
prices = data_transposed[:,1:6]
prices
```

```python
# Graficamos cada precio de los 87 meses para "Moscú"
plt.plot(dates,prices[:,0])
```

```python
city_names = ['Moscu', 'Kaliningrado', 'Petersburgo', 'Krasnodar', 'Ekaterimburgo']
city_prices_dict = {name: prices[:, i] for i, name in enumerate(city_names)}
```

```python
Moscu = city_prices_dict['Moscu']
Kaliningrado = city_prices_dict['Kaliningrado']
Petersburgo = city_prices_dict['Petersburgo']
Krasnodar = city_prices_dict['Krasnodar']
Ekaterimburgo = city_prices_dict['Ekaterimburgo']
```

```python
plt.plot(dates, Moscu)
plt.plot(dates, Kaliningrado)
plt.plot(dates, Petersburgo)
plt.plot(dates, Krasnodar)
plt.plot(dates, Ekaterimburgo)
plt.legend(city_names)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs = axs.flatten()

for i, (city_name, city_prices) in enumerate(city_prices_dict.items()):
    if i < len(axs):
        axs[i].plot(dates, city_prices)
        axs[i].set_title(city_name)
        axs[i].set_xlabel('Months')
        axs[i].set_ylabel('Price')

for j in range(len(city_names), len(axs)):
    fig.delaxes(axs[j])

plt.tight_layout()
plt.show()
```

```python
Moscu - y
```

* Elevar al cuadrado y luego sacar la raíz cuadrada es una forma de calcular la magnitud de un vector, también conocida como la norma o longitud del vector.

```python
np.sum(np.power(Moscu - y, 2))
```

```python
np.sqrt(np.sum(np.power(Moscu - y, 2)))
```

```python
np.linalg.norm(Moscu - y)
```

DEFINIR CON UNA RECTA EL CONJUNTO DE LOS DATOS

```python
n = np.size(Moscu)
y = Moscu
x = dates
```

```python
a = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2) - np.sum(x)**2)
a
```

```python
b = np.mean(y) - a*np.mean(x)
b
```

```python
print(f'La pendiente es: {a}\nLa intersección es: {b}')
```

```python
y = a*x + b
y
```

```python
np.linalg.norm(Moscu - y)
```
