```python
import pandas as pd
import json
import numpy as np
```

```python
data = pd.read_json('/content/datos_hosting.json')
```

```python
data = pd.json_normalize(data['info_inmuebles'])
```

```python
data.head()
```

```python
data.info()
```

```python
columns = list(data.columns)
columns
```

```python
data = data.explode(columns[3:])
data
```

```python
data.reset_index(inplace=True, drop=True)
data
```

```python
data.info()
```

```python
data['max_hospedes'] = data['max_hospedes'].astype(np.int64)
```

```python
num_cols = ['cantidad_baños','cantidad_cuartos','cantidad_camas']
```

```python
data[num_cols] = data[num_cols].astype(np.int64)
```

```python
data['evaluacion_general'] = data['evaluacion_general'].astype(np.float64)
```

```python
data['precio'] = data['precio'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
data['precio'] = data['precio'].astype(np.float64)
```

```python
data[['cuota_deposito','cuota_limpieza']] = data[['cuota_deposito','cuota_limpieza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
data[['cuota_deposito','cuota_limpieza']]
```

```python
data[['cuota_deposito','cuota_limpieza']] = data[['cuota_deposito','cuota_limpieza']].astype(np.float64)
```

```python
data['descripcion_local'] = data['descripcion_local'].str.lower()
data['descripcion_local']
```

```python
data['descripcion_local'][100]
```

```python
data['descripcion_local'] = data['descripcion_local'].str.replace(r'[^a-zA-Z0-9\'\-]',' ',regex=True)
```

```python
data['descripcion_local'] = data['descripcion_local'].str.replace(r'(?<!\w)-(?!\w)',' ',regex=True)
```

```python
data['descripcion_local'].head()
```

```python
# TOKENIZACIÓN -> Lista de palabras
data['descripcion_local'] = data['descripcion_local'].str.split()
data['descripcion_local']
```

```python
data['comodidades'] = data['comodidades'].str.replace(r'\{|}|\"', '', regex=True)
```

```python
data['comodidades'] = data['comodidades'].str.split(',')
data['comodidades']
```

```python
data['descripcion_vecindad'] = data['descripcion_vecindad'].str.lower()
data['descripcion_vecindad'][1]
```

```python
data['descripcion_vecindad'] = data['descripcion_vecindad'].str.replace(r'[^a-zA-Z0-9\'\-]',' ',regex=True)
```

```python
data['descripcion_vecindad'] = data['descripcion_vecindad'].str.replace(r'(?<!\w)-(?!\w)',' ',regex=True)
```

```python
data['descripcion_vecindad'] = data['descripcion_vecindad'].str.split()
data.head()
```
